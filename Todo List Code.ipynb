```python
# Todo list simple coding Approach with Dictionary
#======TODO LIST APP==================
#Main Menu
# Add Tasks
# View Tasks
# Edit Task
# Delete Task
# Reset Tasks
# Terminate App

def isInteger(userInput):
    try:
        int(userInput)
        return True
    except ValueError:
        return False

def MainMenu(menu):
    print("--------------------------\n       Main Menu     \n--------------------------")
    for menu_key, menu_name in menu.items():
        print(f"{menu_key}- {menu_name}")
    userInput = input("\n[-] Enter Command ~ ")
    return userInput;

def AddTask(TaskDictionary):
    print("--------------------------\n|       Adding Task       |\n--------------------------\n")
    while (True):
        taskIndex = len(TaskDictionary) + 1
        
        if taskIndex > 1:
            taskIndex = int(max(TaskDictionary.keys())) + 1
        
        task =input(f"\n[-] Task # {taskIndex} ~ ")
        if task.strip() == "":
            print("\n[-] Empty Task Not Allowed! \n")
            continue
            
        TaskDictionary.update({taskIndex : task})
        print(f"[-] Task # {taskIndex} Added Successfully.\n")
        userInput= input("[-] Do you want to Add Another Task? [Y/N] ~ \n")
        if userInput.lower == "y":
            continue
        else:
            break

def ViewTask(TaskDictionary):
    print("--------------------------\n|      Viewing Task      |\n--------------------------\n")
    if len(TaskDictionary) > 0:
        for (taskIndex, task) in TaskDictionary.items():
            print(f"\n [Task # {taskIndex}] - {task}")
    else:
        print("\n [-] Task Not Found!")
    userInput = input("\n [-] Do you want to Add Tasks? [Y/N] ~ ")
    
    if userInput.lower == "y":
        return 1
    else:
        return 0
        
def DeleteTask(TaskDictionary):
    print("--------------------------\n|      Deleting Task      |\n--------------------------\n")
    DisplayTask = True
    while (True):
        if len(TaskDictionary) > 0:
            if DisplayTask:
                for (taskIndex, task) in TaskDictionary.items():
                    print(f" [Task # {taskIndex}] - {task} ")
                    DisplayTask = False
            taskIndexInput = input("\n [-] Enter Task Index (eg. 1) ~ ")
            
            if isInteger(taskIndexInput):
                taskIndex = int(taskIndexInput)
                if taskIndex in TaskDictionary.keys():
                    TaskDictionary.pop(taskIndex)
                    print(f"\n [-] Task # {taskIndex} Deleted Successfully! ")
                    userInput = input("\n [-] Do you want to delete Another Task? [Y/N] ~ ")
                else:
                    print("\n [-] Invalid Task Index.\n [-] Try Valid Task Index.\n")
                    continue
                if userInput.lower == "y":
                    continue
                else:
                    break
            else:
                print("\n [-] Invalid Task Index.\n [-] Try Valid Index.\n")
                    
        else:
            print("\n[-] Task Not Found! \n")
            userInput = input("[-] Do you want to Add Task? [Y/N] ~ ")
            
            if userInput.lower == 1:
                AddTask(TaskDictionary)
                break
            else:
                break
def AboutDeveloper():
    print("--------------------------\n|      Things To Do      |\n--------------------------\n")
    print("""            This To Do Application is coded by Khansa Ashraf Widhani  
            Contact Info >>>        Email at : khansa.ashraf23@gmail.com 
                                    Facebook : https://www.facebook.com/khansa.ashraf.501
                                    LikedIn  : https://www.linkedin.com/in/khansa-ashraf-4776091a1/
        """)
#Main area            
menu = {            #Menu dictionary
    1 : "Add Task",
    2 : "View Task",
    3 : "Delete Task",
    4 : "Terminate App",
    5 : "About Developer",
    
}
# Empty dictionary for tasks
TaskDictionary = {}

print("--------------------------\n|      Things To Do      |\n--------------------------\n")
name = input("Enter your name: ")
while (True):
    userCommand = MainMenu(menu)
    
    if isInteger(userCommand):
        userCommand = int(userCommand)
        
        #Adding Task
        if userCommand == 1:
            AddTask(TaskDictionary)
            continue
            
        
        #Viewing Tasks
        if userCommand == 2:
            AddTask = ViewTask(TaskDictionary)
            if AddTask == 1:
                AddTask(TaskDictionary)
            else:
                continue
        
        #Deleting Task
        if userCommand == 3:
            DeleteTask(TaskDictionary)
            
        #Terminating App 
        if userCommand == 4:
            print("\n [+] Terminating Application")
            break
        if userCommand == 5:
            AboutDeveloper()
    else:
        print(f"\n[-] Invalid Command. \n Try valid Command between {1} to {len(menu)}\n")
```
