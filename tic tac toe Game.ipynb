```python
#tic - tac - toe Game
#It is coded in a very simpler way. Not used any compex approach. Have Fun :)
from random import randrange
board = [[1,2,3],[4,'X',6],[7,8,9]]
sign = ''
def isInt(user):
    try:
        int(user)
        return True
    except ValueError:
        return False
        
def displayBoard(board):
    print("\n+----------+----------+----------+")
    for fields in range(len(board)):
        #print(fields)
        for boxNum in range(1):
            print("|          |          |          |")
            print("|   ",board[fields][boxNum],"    |    ",board[fields][boxNum + 1],"   |    ",board[fields][boxNum + 2],"   |")
            print("|          |          |          |")
            print("+----------+----------+----------+")

def enterMove(board):
    displayBoard(board)
    if victoryFor(board, sign) == '1':
        loop = True
        while loop:
            freeFields = listOfEmptyFields(board)
            if freeFields == 'Game Tie!':
                break
            else:
                userMove = input("Enter your Move: ")
                if isInt(userMove) == 1:
                    userMove = int(userMove)
                    if userMove >= 1 and userMove <=9:
                        pass
                    else:
                        print("[-] Value Error: Numbers between (1 - 9) are allowed only.")
                        continue
                else:
                    print("[-] Value Error: Numbers are allowed only.")
                    continue
                    
                if userMove in freeFields:
                    if userMove == 1:
                        board[0][0] = 'O'
                    if userMove == 2:
                        board[0][1] = 'O'
                    if userMove == 3:
                        board[0][2] = 'O'
                    if userMove == 4:
                        board[1][0] = 'O'
                    if userMove == 5:
                        board[1][1] = 'O'
                    if userMove == 6:
                        board[1][2] = 'O'
                    if userMove == 7:
                        board[2][0] = 'O'
                    if userMove == 8:
                        board[2][1] = 'O'
                    if userMove == 9:
                        board[2][2] = 'O'
                    loop = False        
                else:
                    print("\n[-] Move out of range. Try valid field")
                    continue
            if not freeFields == 'Game Tie!' :
                drawMove(board)

def listOfEmptyFields(board):
    emptyFields = ()
    for i in board:
        for fields in i:
            if isInt(fields) == 1:
                emptyFields = list(emptyFields)
                emptyFields.append(fields)
                emptyFields = tuple(emptyFields)
    if len(emptyFields) < 1:
        tie = "Game Tie!"
        return tie
    else:
        return emptyFields
    
def victoryFor(board, sign):
    if (board[0][0] == 'X' and board[0][1] == 'X' and board[0][2] == 'X') or (board[0][0] == 'X' and board[1][0] == 'X' \
    and board[2][0] == 'X')or (board[2][0] == 'X' and board[2][1] == 'X' and board[2][2] == 'X') or (board[0][2] == 'X' \
    and board[1][2] == 'X' and board[2][2] == 'X'):
        sign = 'draw'
        
    elif (board[0][0] == 'O' and board[0][1] == 'O' and board[0][2] == 'O') or (board[0][0] == 'O' and board[1][0] == 'O' \
    and board[2][0] == 'O')or (board[2][0] == 'O' and board[2][1] == 'O' and board[2][2] == 'O') or (board[0][2] == 'O'and\
    board[1][2] == 'O' and board[2][2] == 'O'):
        sign = 'player'
    else:
        sign = '1'
    if sign == '1':
        return sign
    elif sign == 'draw':
        print("[+] Oops, You Lose!")
    else:
        print("[+] Congo! You Won")
def drawMove(board): #computer move
    displayBoard(board)
    if victoryFor(board, sign) == '1':
        loop = True
        while loop:
            draw = randrange(1,10)
            freeFields = listOfEmptyFields(board)
            if draw in freeFields:
                if draw == 1:
                    board[0][0] = 'X'
                if draw == 2:
                    board[0][1] = 'X'
                if draw == 3:
                    board[0][2] = 'X'
                if draw == 4:
                    board[1][0] = 'X'
                if draw == 5:
                    board[1][1] = 'X'
                if draw == 6:
                    board[1][2] = 'X'
                if draw == 7:
                    board[2][0] = 'X'
                if draw == 8:
                    board[2][1] = 'X'
                if draw == 9:
                    board[2][2] = 'X'
                loop = False
            else:
                continue
        enterMove(board)
enterMove(board)
```
